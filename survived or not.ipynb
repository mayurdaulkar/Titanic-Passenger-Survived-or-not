```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
# importing the libraries
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df = df.drop(columns = ['PassengerId' , 'Name', 'SibSp' , 'Parch' , 'Ticket' , 'Cabin' , 'Embarked'])
# dropping the unnecessary columns 
```

```python
df.head()
```

```python
target = df.Survived
# setting the target
```

```python
df= pd.get_dummies(df, columns=['Sex'])
# setting the dummies so make a separate column for male and female 
```

```python
df.head()
```

```python
df['Sex_female'] = df['Sex_female'].astype(int)
df['Sex_male'] = df['Sex_male'].astype(int)
# converting the boolean to the int 
```

```python
df.head()
```

```python
df1 = df.drop('Survived' , axis = 'columns')
```

```python
df1.isnull().sum()
# checking for the null values
```

```python
e1 = df.Age.mean()
# mean of Age to fill into the empty rows
```

```python
e1
```

```python
df1.Age = df1.Age.fillna(e1)
# filling the null values of the Age
```

```python
df1.isnull().sum()
#again checking for null values
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(df1 , target , test_size = 0.2)
# splitting the data for the training and testing purpose
```

```python
from sklearn.naive_bayes import GaussianNB
reg = GaussianNB()
#model training
```

```python
reg.fit(X_train , Y_train)
# fitting the training data into the model 
```

```python
reg.score(X_test , Y_test)
# checking the score of the prediction 
```

```python
ans = reg.predict([[3,	35.0	,8.0500,	0,	1]])
if(ans[0]==0):
    print('the person is not survived')
else:
    print('The person is survived')
# making random predictions and checking if the person is survived or not
```
